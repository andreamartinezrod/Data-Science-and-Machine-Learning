```python
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt

#Primero definimos la funci√≥n a la cual queremos calcular la integral
```

```python
def f(x):
    return np.exp(-x**2)
```

```python
x = np.linspace(-50, 50, 200)
y = f(x)
plt.plot(x, y)
```

```python
resultado = integrate.quad(f,0,1)
resultado
```

```python
x_values = np.linspace(0,1,100)
y_values = []
for x in x_values:
    y_values.append(f(x))
    plt.plot(x_values, y_values)

resultado_2 = np.trapezoid(x_values,y_values)
```

```python
from  scipy.interpolate import interp1d

x = [0,1,2,3,4]
y = [1,2.7,5.8,6.6,7.5]
f1 = interp1d(x,y)
```

```python
print("El valor de f(2.5) es: ", f1(2.5))
```

```python
plt.plot(x, y, 'o')
plt.plot(x, f1(x), '--', color='red')
plt.plot(2.5,f1(2.5), 'o', color='red')
plt.show()
```
