El análisis de datos es un proceso fundamental en la ciencia moderna y en la toma de decisiones basada en evidencia. Consiste en inspeccionar, limpiar, transformar y modelar datos con el objetivo de descubrir información útil, llegar a conclusiones relevantes y respaldar la toma de decisiones.
Este proceso es aplicable a una amplia gama de disciplinas, como la ciencia, la industria, los negocios, la salud y el gobierno, donde los datos son cada vez más abundantes y complejos.  


El análisis de datos se puede abordar desde distintos enfoques. Por un lado, existen métodos cuantitativos, como los modelos estadísticos y matemáticos, que buscan representar fenómenos de forma estructurada. Por otro lado, están los enfoques más exploratorios o empíricos, que permiten descubrir patrones ocultos sin hipótesis previas mediante técnicas como el análisis visual o los algoritmos de aprendizaje automático.  

Un componente esencial del análisis de datos es la inferencia estadística, la cual permite tomar decisiones bajo incertidumbre. A través del uso de probabilidades, se pueden estimar parámetros poblacionales, probar hipótesis y predecir comportamientos futuros, todo ello con un nivel de confianza determinado.  

El proceso de análisis también implica entender el contexto del problema, definir correctamente las preguntas de investigación, seleccionar y preparar los datos adecuados, y finalmente interpretar los resultados obtenidos. Además, la comunicación efectiva de los hallazgos, mediante visualizaciones claras y argumentos bien fundamentados, es clave para generar impacto y facilitar la toma de decisiones informadas.  



### Tipos
El análisis de datos puede clasificarse según diferentes criterios, principalmente el tipo de información que se estudia y el objetivo del análisis.  

1. Según el tipo de información:  
* Análisis cuantitativo: Se enfoca en estudiar datos que pueden medirse numéricamente. Este tipo de análisis busca relaciones, patrones y tendencias utilizando herramientas estadísticas, matemáticas y computacionales. Es común en estudios experimentales, encuestas, mediciones financieras, entre otros.

* Análisis cualitativo: Se centra en información no numérica, como textos, entrevistas, observaciones o fenómenos sociales. Busca interpretar significados, contextos, experiencias o comportamientos humanos. Este enfoque es común en ciencias sociales, antropología, educación y estudios culturales.  

2. Según el objetivo del análisis:  
* Análisis exploratorio: Su propósito principal es descubrir patrones, estructuras o relaciones entre los datos sin una hipótesis previa. Se utiliza al inicio de un proyecto para comprender mejor la naturaleza de los datos y generar nuevas preguntas o hipótesis. Emplea herramientas como gráficos, estadísticos descriptivos y técnicas de reducción de dimensión (como PCA).  
* Análisis confirmatorio: Tiene como objetivo probar hipótesis previamente formuladas. Utiliza técnicas estadísticas formales para verificar si existe suficiente evidencia en los datos que respalde una afirmación. Es común en investigaciones científicas donde se quiere validar o refutar teorías.

## Conversión de fechas en cadenas con diferentes formatos

d.strftime(formato) : Devuelve la cadena que resulta de transformar la fecha d con el formato indicado en la cadena formato. La cadena formato puede contener los siguientes marcadores de posición: %Y (año completo), %y (últimos dos dígitos del año), %m (mes en número), %B (mes en palabra), %d (día), %A (día de la semana), %a (día de la semana abrevidado), %H (hora en formato 24 horas), %I (hora en formato 12 horas), %M (minutos), %S (segundos), %p (AM o PM), %C (fecha y hora completas), %x (fecha completa), %X (hora completa).

Como vimos, en Python, se puede convertir un objeto de tipo datetime a una cadena de texto utilizando el método strftime(formato). Este método devuelve una representación en texto de la fecha según el formato indicado. Los formatos se definen mediante códigos especiales que comienzan con %, cada uno representando un componente de la fecha u hora.   

**Principales códigos de formato de strftime**

* %Y: Año completo en 4 dígitos (por ejemplo, 2025)

* %y: Últimos dos dígitos del año (por ejemplo, 25)

* %m: Mes en número, con dos dígitos (01 a 12)

* %B: Nombre completo del mes (por ejemplo, July)

* %b: Nombre abreviado del mes (por ejemplo, Jul)

* %d: Día del mes con dos dígitos (01 a 31)

* %A: Nombre completo del día de la semana (por ejemplo, Wednesday)

* %a: Nombre abreviado del día (por ejemplo, Wed)

* %H: Hora en formato 24 horas (00 a 23)

* %I: Hora en formato 12 horas (01 a 12)

* %M: Minutos (00 a 59)

* %S: Segundos (00 a 59)

* %p: AM o PM

* %x: Representación local de la fecha

* %X: Representación local de la hora

* %c: Fecha y hora completas en formato local

# End-to-End  ciencia de datos

El desarrollo de un proyecto de ciencia de datos típicamente sigue una serie de pasos estructurados:  

1. Comprensión del problema (Look at the big picture):
Se define el objetivo del proyecto, comprendiendo el contexto del negocio o área de aplicación, para formular una **meta clara y medible**.

2. Obtención de los datos (Get the data):
Se recopilan los datos necesarios desde diversas fuentes, ya sean bases internas, APIs, sensores o scraping.

3. Exploración y visualización (Explore and visualize the data):
Se realiza un análisis exploratorio para entender la estructura, distribución, correlaciones y posibles anomalías en los datos. La visualización ayuda a descubrir patrones clave.

4. Preparación de los datos (Prepare the data):
Se limpian, transforman y escalan los datos. Esto incluye tratar valores faltantes, convertir variables categóricas y dividir conjuntos de entrenamiento y prueba.

5. Selección y entrenamiento del modelo (Select a model and train it):
Se eligen modelos adecuados y se entrenan con los datos preparados. Esto puede incluir regresión, árboles de decisión, redes neuronales, etc.

6. Ajuste fino (Fine-tune your model):
Se optimizan los hiperparámetros y se evalúa el desempeño usando validación cruzada o pruebas A/B para mejorar la precisión.

7. Presentación de la solución (Present your solution):
Se comunican los resultados y hallazgos a las partes interesadas mediante visualizaciones, reportes o dashboards interactivos.

8. Despliegue y mantenimiento (Launch, monitor, and maintain):
El modelo se integra en un sistema real, se monitorea su rendimiento en producción y se actualiza conforme cambian los datos o los objetivos del negocio.

# Selección de métrica de evaluación 

Una métrica de evaluación en machine learning es una medida cuantitativa que permite valorar qué tan bien un modelo está funcionando, comparando sus predicciones con los valores reales.

**Es importante porque:**

* Guía la selección y ajuste del modelo, ayudando a identificar cuál es más preciso o adecuado para el problema.

* Permite cuantificar el desempeño de forma objetiva, facilitando comparaciones entre diferentes modelos o configuraciones.

* Ayuda a detectar problemas como sobreajuste o sesgos en los datos.

* Sirve para tomar decisiones informadas sobre la viabilidad y confiabilidad del modelo antes de desplegarlo en producción.  

Sin una métrica clara, no es posible saber si el modelo realmente cumple con los objetivos esperados.

Es importante elegir una métrica de desempeño para evaluar qué tan bien funciona un modelo. En problemas de regresión, una métrica común es el error cuadrático medio raíz (RMSE), que mide el error promedio en las predicciones, penalizando más los errores grandes. Esto ayuda a cuantificar la precisión del modelo y guiar su mejora.  

Aunque el RMSE es la métrica más común para problemas de regresión, en algunos casos, como cuando hay muchos valores atípicos, puede ser mejor usar otra métrica como el error absoluto medio (MAE), que es menos sensible a esos valores extremos.

Muchas métricas de evaluación en machine learning se basan en funciones de distancia, porque miden qué tan “lejos” están las predicciones del modelo de los valores reales. Esto permite cuantificar el error o la diferencia entre ambos.

Las métricas de distancia son funciones matemáticas que calculan una medida de discrepancia o separación entre dos puntos o vectores en un espacio — en este caso, entre los valores predichos y los valores reales.

### Tipos comunes de métricas de distancia en machine learning  


* Error absoluto (L1) o Mean Absolute Error (MAE): Suma o promedio de las diferencias absolutas entre predicción y realidad.Penaliza los errores linealmente y es robusto a valores atípicos.

* Error cuadrático (L2) o Mean Squared Error (MSE): Promedio de los cuadrados de las diferencias. Penaliza errores grandes más severamente.

* Root Mean Squared Error (RMSE): Raíz cuadrada del MSE. Tiene las mismas unidades que la variable objetivo, lo que facilita interpretación.

* Distancia de Minkowski: Generaliza L1 y L2 con un parámetro que controla la “potencia” de la distancia.


* Distancia coseno: Mide el ángulo entre vectores, usada comúnmente en texto y recomendaciones.

El método anterior, es un método de muestreo aleatorio puro, lo cual funciona bien cuando el conjunto de datos es suficientemente grande. Sin embargo, si no es así, existe el riesgo de introducir un sesgo de muestreo significativo. Por ejemplo, en una encuesta telefónica, no se eligen personas al azar sin más, sino que se busca que la muestra sea representativa de la población en aspectos importantes, como el género. Esto se llama muestreo estratificado, donde la población se divide en subgrupos homogéneos (estratos) y se seleccionan muestras adecuadas de cada uno para asegurar representatividad.

Si se muestreara puramente al azar, podría haber una alta probabilidad de obtener una muestra sesgada, lo que afectaría la validez de los resultados.

En el contexto del análisis de precios de viviendas, si se sabe que el ingreso medio es una variable clave, es importante que el conjunto de prueba refleje bien las distintas categorías de ingreso. ### ¿Qué es el coeficiente de Pearson?
El coeficiente de correlación de Pearson (también llamado Pearson’s r) es una medida estadística que indica la fuerza y dirección de la relación lineal entre dos variables numéricas.

Su valor está entre –1 y 1:  

* +1: correlación lineal positiva perfecta (cuando una variable sube, la otra también).

* 0: no hay correlación lineal.

* –1: correlación lineal negativa perfecta (cuando una sube, la otra baja).

#### Métodos disponibles en df.corr(method=...):
1. 'pearson' (por defecto)
* Mide: Correlación lineal.

* Requiere: Datos numéricos y continuos.

* Supone: Relación lineal y distribución normal.

* Escala: De –1 a 1.

2. 'kendall'
* Mide: Concordancia entre rangos.

* Útil cuando: Hay muchos valores atípicos o datos no normales.

* Más robusto, pero más lento en grandes conjuntos de datos.

* Tipo: Correlación no paramétrica (por rangos).

3. 'spearman'
* Mide: Correlación entre rangos (valores ordenados).

* Detecta relaciones monótonas (no necesariamente lineales).

* También resistente a valores atípicos.

* Tipo: Correlación no paramétrica

### One Hot Encoder

El One Hot Encoder es una técnica para convertir variables categóricas en variables numéricas binarias que los modelos de machine learning pueden entender mejor.  

En vez de asignar un número entero a cada categoría (como hace el Ordinal Encoder), el One Hot Encoder crea una columna binaria para cada categoría posible. En cada fila, solo una de esas columnas tendrá un valor 1 (indicando la presencia de esa categoría) y el resto serán 0.  

¿Por qué es bueno?

Evita que el modelo interprete un orden o relación numérica entre categorías que en realidad no existe (por ejemplo, que la categoría 2 sea “mayor” que la 1).  

Representa las categorías de forma clara y sin sesgos, ayudando a que modelos lineales y muchos otros funcionen correctamente con variables categóricas.

#### One Hot Encoder vs get_dummies():

* get_dummies() es una función de pandas que también crea variables dummy (binarias) para categorías, muy útil y rápida para convertir columnas categóricas a variables numéricas.

* ¿Por qué One Hot Encoder de scikit-learn puede ser mejor?

1. Integración con pipelines y transformaciones: One Hot Encoder se integra perfectamente con el flujo de trabajo de scikit-learn, como pipelines, lo que permite hacer transformaciones consistentes en entrenamiento y prueba, sin fugas de datos (data leakage).

2. "Recuerda" las categorías: Esto significa que cuando entrenas el encoder, él aprende cuáles son las categorías presentes en el conjunto de entrenamiento. Luego, cuando aplicas la transformación a nuevos datos (por ejemplo, en el conjunto de prueba o producción), si aparece una categoría nueva o faltante, el encoder sabe cómo manejarla (puede ignorarla o manejarla sin errores). En cambio, get_dummies() no “recuerda” categorías porque cada llamada es independiente y no guarda estado.

3. Más control: One Hot Encoder permite manejar valores desconocidos con el parámetro handle_unknown='ignore', cosa que get_dummies() no hace fácilmente.

# Escalamiento y transformaciones

El escalado de características es una transformación crucial en la ciencia de los datos y particularmente en los modelos de ML, ya que la mayoría de los algoritmos funcionan mal cuando las variables numéricas tienen escalas muy diferentes. Por ejemplo, en los datos de viviendas, el número total de habitaciones varía entre 6 y 39,320, mientras que los ingresos medios van de 0 a 15. Sin escalado, los modelos tienden a dar más importancia a las variables con mayor rango, ignorando otras importantes. Las dos técnicas más comunes para unificar las escalas son el escalado min-max y la estandarización.

### Normalización  
Consiste en escalar los datos para que sus valores queden dentro de un rango específico, comúnmente entre 0 y 1. Esto se logra restando el valor mínimo y dividiendo entre el rango (máximo - mínimo).  

La normalización es más útil cuando no se conoce la distribución o se quiere restringir los datos a un rango específico, por ejemplo para algoritmos basados en distancias (como KNN o redes neuronales) o cuando se necesita garantizar que todas las características estén en la misma escala.  

Se prefiere cuando los datos no tienen distribución normal o cuando el modelo se basa en distancias o requiere un rango específico (redes neuronales, KNN).

### Estandarización  
Es un proceso que transforma los datos para que tengan media 0 y desviación estándar 1. Esto se hace restando la media y dividiendo por la desviación estándar de la variable. El resultado es que los datos quedan centrados y con varianza uniforme.  

La estandarización mantiene la forma de la distribución y es útil cuando los datos tienen distribución aproximadamente normal o cuando los algoritmos asumen datos centrados (por ejemplo, regresión lineal, SVM, PCA).

Se prefiere cuando los datos tienen distribución normal o cuando el modelo es sensible a la media y varianza (regresión lineal, SVM, PCA).  

Es menos sensible a outliers

Es fundamental ajustar (fit) los escaladores únicamente con los datos de entrenamiento; nunca se debe usar fit() o fit_transform() en otros conjuntos como validación o prueba. ## Transformación 

Cuandoo una variable tiene una distribución con "cola pesada" (es decir, muchos valores alejados de la media que no son poco frecuentes de forma exponencial), tanto la normalización (min-max) como la estandarización pueden aplastar la mayoría de los valores en un rango muy pequeño. Esto es problemático porque los modelos de machine learning generalmente no funcionan bien con esas distribuciones.

Por eso, antes de escalar esa variable, es recomendable transformarla para reducir esa cola pesada y, si es posible, hacer que su distribución sea más simétrica. Por ejemplo, para variables positivas con cola pesada hacia la derecha, una técnica común es usar la raíz cuadrada o elevar la variable a una potencia entre 0 y 1. Si la cola es muy larga y pesada, como en distribuciones de ley de potencias (power law), usar el logaritmo de la variable puede ayudar.

Si tomamos la variable "población", vemos que sigue una ley de potencias, donde los distritos con 10,000 habitantes son solo 10 veces menos frecuentes que los de 1,000 habitantes (no exponencialmente menos). Al aplicar el logaritmo a esta variable, su distribución se acerca mucho a una distribución normal o campana (Gaussiana), lo cual es más favorable para los modelos.

## bucketizing

Otra forma de manejar variables con distribuciones heavy-tailed (colas pesadas) es hacer bucketizing, que consiste en dividir la distribución de esa variable en grupos o "buckets" de tamaño aproximadamente igual. Luego, cada valor original se reemplaza por el índice del bucket al que pertenece, similar a cómo se creó la variable income_cat para muestreo estratificado.

Por ejemplo, se puede reemplazar cada valor por su percentil. Al hacer esto con buckets de igual tamaño, la nueva variable resultante tendrá una distribución casi uniforme, lo que elimina la necesidad de hacer un escalado adicional. Además, si se quiere, se puede normalizar dividiendo los índices de los buckets por el número total de buckets para que los valores queden en un rango de 0 a 1.

```python
# Supongamos que tienes un DataFrame housing_num con la columna population
population = housing_num['population']

# Creamos 10 buckets (deciles) usando pd.qcut para dividir en percentiles iguales
housing_num['population_bucket'] = pd.qcut(population, q=10, labels=False)

# Ahora cada valor en 'population_bucket' es un número de 0 a 9 que indica el bucket
print(housing_num[['population', 'population_bucket']].head())

# Si quieres que los valores estén en rango [0,1], divide por número de buckets
housing_num['population_bucket_normalized'] = housing_num['population_bucket'] / 9

print(housing_num[['population_bucket', 'population_bucket_normalized']].head())
```

```python
plt.figure(figsize=(12,8))
plt.hist(housing_num['population_bucket'],bins=20,edgecolor="black",color="green",alpha=.5,
         label="population_bucket")

plt.legend()
plt.title("Population",size=20)
plt.grid()
plt.xlim(2,10)
plt.show()
```

Aunque el enfoque anterior funciona correctamente, existe una alternativa más sencilla: utilizar TransformedTargetRegressor. Esta clase nos permite encapsular tanto el modelo de regresión como el transformador de etiquetas en un solo objeto. Solo necesitamos crearla pasando como argumentos el modelo y el transformador que queremos aplicar a las etiquetas, y luego ajustarla directamente con los datos de entrenamiento, utilizando las etiquetas originales sin escalar.

Internamente, TransformedTargetRegressor se encarga de transformar las etiquetas antes de entrenar el modelo, y también de revertir esa transformación al momento de predecir. # Python pipelines

Un pipeline es una herramienta que permite encadenar varias etapas del proceso de modelado en un solo objeto estructurado. Cada etapa puede ser un transformador (como escaladores, imputadores o generadores de características) o un modelo de aprendizaje. El objetivo es que los datos pasen automáticamente de una etapa a la siguiente, en el orden especificado.

Make_pipeline es una función de Scikit-Learn que permite construir de forma sencilla un pipeline (flujo de trabajo encadenado) en el que se aplican varios pasos de procesamiento secuencialmente, como transformación de datos y entrenamiento de un modelo.  
Su principal utilidad es organizar el código de forma clara y reproducible, encapsulando todo el proceso (por ejemplo, escalado de datos seguido de una regresión) en un solo objeto. Esto facilita tareas como validación cruzada, predicciones o serialización del modelo completo.  

Su uso es simple: se pasan como argumentos los pasos deseados (por ejemplo, make_pipeline(StandardScaler(), LinearRegression())), y el pipeline se encargará automáticamente de aplicar cada paso en orden al llamar a fit(), predict() o score(). Esta herramienta es especialmente valiosa en proyectos de machine learning, ya que reduce errores, mantiene consistencia entre entrenamiento y prueba, y permite integrar todo el flujo en evaluaciones automáticas como GridSearchCV.

```python
from sklearn.pipeline import make_pipeline
```

```python
num_pipeline = make_pipeline(SimpleImputer(strategy="median"), StandardScaler())
```

Cuando se invoca el método fit() de un pipeline, este aplica internamente fit_transform() a cada uno de los transformadores en orden, utilizando la salida de cada uno como entrada para el siguiente. Este proceso continúa hasta llegar al último paso, que debe ser un estimador final; a este simplemente se le aplica el método fit().  

El pipeline hereda los métodos del estimador final. Por ejemplo, si el último elemento es un StandardScaler, que es un transformador, entonces el pipeline completo también funcionará como transformador y podrá utilizarse con transform(). Al llamar a transform(), el pipeline aplicará de manera secuencial todas las transformaciones definidas. Por otro lado, si el último paso fuera un modelo de predicción (como un clasificador o regresor), el pipeline tendría un método predict() en lugar de transform(), y al usarlo, primero ejecutaría todas las transformaciones previas y luego llamaría a predict() con los datos ya procesados.

```python
housing_num_prepared = num_pipeline.fit_transform(housing_num)
```

Hasta ahora hemos procesado por separado las columnas numéricas y categóricas, pero lo ideal sería contar con un único transformador que pueda aplicar el tratamiento adecuado a cada tipo de columna dentro de un mismo flujo. Para lograr esto, se puede utilizar ColumnTransformer, una herramienta que permite asignar distintas transformaciones a diferentes columnas de forma organizada. Por ejemplo, podemos usar un ColumnTransformer que aplique num_pipeline a las variables numéricas y cat_pipeline a las categóricas, automatizando así el preprocesamiento completo de nuestros datos en un solo objeto.



```python
from sklearn.compose import ColumnTransformer
```

```python
num_attribs = ["longitude", "latitude", "housing_median_age", "total_rooms",
"total_bedrooms", "population", "households", "median_income"]
```

```python
cat_attribs = ["ocean_proximity"]
```

```python
cat_pipeline = make_pipeline(
                SimpleImputer(strategy="most_frequent"),
                OneHotEncoder(handle_unknown="ignore"))
```

Preprocesamiento completo

```python
preprocessing = ColumnTransformer([("num", num_pipeline, num_attribs),
                                    ("cat", cat_pipeline, cat_attribs),
                                    ])
```

```python
preprocessing
```

```python
pd.DataFrame(preprocessing.fit_transform(housing),columns= preprocessing.get_feature_names_out(),
            index= housing.index)
```

### Más eficiente

```python
from sklearn.compose import make_column_selector, make_column_transformer
```

```python
preprocessing = ColumnTransformer([("num", num_pipeline, make_column_selector(dtype_include=np.number)),
                                    ("cat", cat_pipeline, make_column_selector(dtype_include=object)),
                                    ])
```

```python
preprocessing
```

```python
pd.DataFrame(preprocessing.fit_transform(housing),columns= preprocessing.get_feature_names_out(),
            index= housing.index)
```
